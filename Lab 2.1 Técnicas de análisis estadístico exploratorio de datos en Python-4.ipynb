```python
# Importe mínimo de ventas
min_ventas = np.min(ventas_diarias)

# Cálculo del rango intercuartílico (IQR)
Q1 = np.percentile(ventas_diarias, 25)
Q3 = np.percentile(ventas_diarias, 75)
IQR = Q3 - Q1

# Cálculo de la varianza
varianza = np.var(ventas_diarias)

# Cálculo de la desviación típica
desviacion_tipica = np.std(ventas_diarias)

min_ventas, IQR, varianza, desviacion_tipica
```

<div style="background-color:#F7DC6F"><b>Resolución Ejercicio 03</b></div>

<div style="background-color:#5DADE2"><b>Apartado 1</b></div>

```python
import pandas as pd

# Tabla de frecuencias absolutas y relativas para 'edad'
frecuencia_edad_abs = edad.value_counts()
frecuencia_edad_rel = edad.value_counts(normalize=True)

# Mostrar resultados
print("Frecuencia Absoluta de Edad:")
print(frecuencia_edad_abs)
print("\nFrecuencia Relativa de Edad:")
print(frecuencia_edad_rel)

# Tabla de frecuencias absolutas y relativas para 'deporte'
frecuencia_deporte_abs = deporte.value_counts()
frecuencia_deporte_rel = deporte.value_counts(normalize=True)

# Mostrar resultados
print("\nFrecuencia Absoluta de Deporte:")
print(frecuencia_deporte_abs)
print("\nFrecuencia Relativa de Deporte:")
print(frecuencia_deporte_rel)

# Tabla de frecuencias absolutas y relativas para 'habilidad'
frecuencia_habilidad_abs = habilidad.value_counts()
frecuencia_habilidad_rel = habilidad.value_counts(normalize=True)

# Mostrar resultados
print("\nFrecuencia Absoluta de Habilidad:")
print(frecuencia_habilidad_abs)
print("\nFrecuencia Relativa de Habilidad:")
print(frecuencia_habilidad_rel)

```

```python
# Crear la tabla de frecuencias absolutas de deporte vs habilidad
tabla_frecuencias_abs = pd.crosstab(deporte, habilidad)

# Crear la tabla de frecuencias relativas
tabla_frecuencias_rel = pd.crosstab(deporte, habilidad, normalize=True)

# Mostrar resultados
print("\nTabla de Frecuencias Absolutas:")
print(tabla_frecuencias_abs)
print("\nTabla de Frecuencias Relativas:")
print(tabla_frecuencias_rel)
```

```python
# Calcular moda, media, mediana, rango y varianza para 'edad'
moda_edad = edad.mode()
media_edad = edad.mean()
mediana_edad = edad.median()
rango_edad = edad.max() - edad.min()
varianza_edad = edad.var()

print("\nModa de Edad:", moda_edad)
print("Media de Edad:", media_edad)
print("Mediana de Edad:", mediana_edad)
print("Rango de Edad:", rango_edad)
print("Varianza de Edad:", varianza_edad)

#Para las variables deporte y habilidad, que son categóricas, solo calculamos la moda:
# Moda de deporte y habilidad
moda_deporte = deporte.mode()
moda_habilidad = habilidad.mode()

print("\nModa de Deporte:", moda_deporte)
print("Moda de Habilidad:", moda_habilidad)
```

```python
def varianza_custom(datos):
    n = len(datos)
    media = sum(datos) / n
    return sum((x - media)**2 for x in datos) / (n - 1)

# Aplicar la función sobre 'edad' y comparar con np.var
import numpy as np
print("\nVarianza con función personalizada:", varianza_custom(edad))
print("Varianza con np.var:", np.var(edad, ddof=1))
```

El objetivo es que repliquéis la función [`summary` predefinida en R](https://www.rdocumentation.org/packages/base/versions/3.6.2/topics/summary) y lo apliquéis al conjunto de datos que se os proporciona.
```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

# Crear el dataframe
data_df = pd.DataFrame({
    'Sexo': ["Mujer", "Hombre", "Mujer", "Mujer", "Mujer", "Hombre", "Mujer", "Hombre", "Hombre", "Mujer", "Mujer", "Hombre", "Hombre", "Mujer", "Mujer", "Hombre", "Mujer", "Mujer", "Mujer", "Mujer"],
    'Edad': [21, 19, 18, 20, 23, 22, 22, 20, 21, 21, 22, 20, 22, 19, 19, 21, 20, 21, 22, 23],
    'Estatura': [1.82, 1.83, 1.78, 1.79, 1.80, 1.90, 1.79, 1.83, 1.72, 1.65, 1.73, 1.79, 1.80, 1.77, 1.69, 1.75, 1.66, 1.59, 1.79, 1.80],
    'Puntuación': [5, 6, 7, 5, 9, 7, 8, 3, 9, 2, 5, 8, 7, 6, 3, 4, 5, 6, 2, 8],
    'Calificación': ["Aprobado", "Aprobado", "Notable", "Aprobado", "Sobresaliente", "Notable", "Notable", "Suspenso", "Sobresaliente", "Suspenso", "Aprobado", "Notable", "Notable", "Aprobado", "Suspenso", "Suspenso", "Aprobado", "Aprobado", "Suspenso", "Notable"]
})

# Filtramos solo las columnas numéricas
data_numeric = data_df[['Edad', 'Estatura', 'Puntuación']]

# Mostramos el dataframe con las variables numéricas
data_numeric
```

```python
#Cálculo de los estadísticos descriptivos:
def summary_stats(df):
    summary = {}

    for column in df.columns:
        data = df[column]

        # Medidas de tendencia central
        mode = data.mode()[0]  # El primer valor más frecuente
        mean = data.mean()  # Media
        median = data.median()  # Mediana

        # Medidas de posición
        q1 = data.quantile(0.25)  # Primer cuartil
        q3 = data.quantile(0.75)  # Tercer cuartil
        percentiles = [data.quantile(p / 100) for p in range(0, 101, 10)]  # Percentiles 0, 10, ..., 100
        deciles = [data.quantile(d / 10) for d in range(0, 11)]  # Deciles 0, 1, ..., 10

        # Medidas de dispersión
        range_ = data.max() - data.min()  # Rango
        iqr = q3 - q1  # Rango intercuartílico
        variance = data.var()  # Varianza
        std_dev = data.std()  # Desviación estándar

        # Medidas de forma
        skewness = skew(data)  # Asimetría
        kurt = kurtosis(data)  # Curtosis

        # Almacenar los resultados
        summary[column] = {
            'Modo': mode,
            'Media': mean,
            'Mediana': median,
            'Q1': q1,
            'Q3': q3,
            'Percentiles': percentiles,
            'Deciles': deciles,
            'Rango': range_,
            'Rango Intercuartílico': iqr,
            'Varianza': variance,
            'Desviación Típica': std_dev,
            'Asimetría': skewness,
            'Curtosis': kurt
        }

    return summary

# Aplicamos la función
summary_results = summary_stats(data_numeric)

# Mostramos los resultados
summary_results
```
