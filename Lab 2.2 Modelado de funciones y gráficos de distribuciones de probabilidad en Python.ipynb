### Ejercicio 1

Supongamos que gestionamos una pequea empresa que vende un producto online. Basndonos en histricos, sabemos que, en promedio, el 20% de las personas que visitan nuestro sitio web realizan una compra. Ahora, en un da particular, tenemos 500 visitantes en el sitio. Asumir que la variable aleatoria $x$ (nmero de compras) sigue una distribucin binomial, ya que tenemos un nmero fijo de ensayos (500 visitantes), cada ensayo es independiente, y cada ensayo resulta en xito (compra) con una probabilidad del 20%.

1. Qu distribucin de probabilidad tiene $x$? Tenis que justificar el tipo de distribucin y sus parmetros.

2. Cul es la probabilidad de tener exactamente 100 compras en un da, dado que 500 personas visitan el sitio?

3. Cul es la probabilidad de tener ms de 100 compras en un da, dado que 500 personas visitan el sitio?

4. Cul es la probabilidad de tener menos de 100 compras en un da, dado que 500 personas visitan el sitio?

5. Representar grficamente la funcin de masa de la variable aleatoria $x$ e interpretar los resultados.

<div style="background-color:#F7DC6F"><b>Resolucin Ejercicio 01</b></div>

<div style="background-color:#5DADE2"><b>Apartado 1</b></div>

```python
#1. Qu distribucin de probabilidad tiene $x$? Tenis que justificar el tipo de distribucin y sus parmetros.
La funcin a continuacin es una funcin Binomial (X ~ B(500, 0.20))
Por los siguientes motivos:
* Tenemos un nmero fijo de ensayos (500)
* Cada ensayo es independiente
* Solo hay dos resultados posibles: compra o no compra
* La probabilidad de xito es constante (20%)
```

<mark>Escribe tu resumen y razonamiento de la solucin alcanzada en el apartado anterior aqu<mark>

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

```python
#2. Cul es la probabilidad de tener exactamente 100 compras en un da, dado que 500 personas visitan el sitio?
from scipy.stats import binom

#Parmetros
n = 500
p = 0.2

#Creamos la distribucin binomial
bin_compras = binom(n=n, p=p)

#Queremos calcular la probabilidad de tener exactamente 100 compras en un da
prob_eq_100 = bin_compras.pmf(100)
print(f'La probabilidad de tener exactamente 100 compras en un da: {prob_eq_100}')
```

<mark>Escribe tu resumen y razonamiento de la solucin alcanzada en el apartado anterior aqu<mark>

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

```python
#3. Cul es la probabilidad de tener ms de 100 compras en un da, dado que 500 personas visitan el sitio?
from scipy.stats import binom

#Parmetros
n= 500
p = 0.2

#Creamos la distribucin binomial
bin_compras = binom(n=n, p=p)

prob_h_100 = 1 - bin_compras.cdf(100)
print(f'La probabilidad de tener ms de 100 compras en un da: {prob_h_100}')
```

```python
#4. Cul es la probabilidad de tener menos de 100 compras en un da, dado que 500 personas visitan el sitio?
from scipy.stats import binom

#Parmetros
n= 500
p= 0.2

#Creamos la distribucin binmomial
bin_compras = binom(n=n, p=p)

prob_l_100 = bin_compras.cdf(99)
print(f'La probabilidad de tener menos de 100 compras en un da (es decir 99): {prob_l_100}')
```

### Ejercicio 2

En una lnea de produccin de chips para ordenadores, se sabe que hay una probabilidad del 5% de que un chip sea defectuoso. 1. Qu distribucin de probabilidad consideramos en este caso? Tenis que justificar el tipo de distribucin y sus parmetros.

4. Generar una muestra de tamao $1000$ de una distribucin geomtrica siendo $p=0.05$ y luego calcular las frecuencias relativas (`pd.Series().count_values(normalize=True)`) y compararlas con las probabilidades tericas (`.pmf()`).

```python
#1. Qu distribucin de probabilidad consideramos en este caso? Tenis que justificar el tipo de distribucin y sus parmetros.
La funcin a continuacin es una funcin Geomtrica con parmetro p= 0.05
Por los siguientes motivos:
* Contamos el nmero de ensayos hasta el primer xito
* Cada ensayo es independiente
* La probabilidad de xito es constante
* No hay lmite en el nmero de ensayos
```

```python
#2. Calcular la probabilidad de que se necesiten probar 10 chips para encontrar el primer chip defectuoso. 
from scipy.stats import geom

#Parmetros
k = 9
p = 0.05

#Creamos la distribucin geomtrica
geom_defecto = geom(p=p)

#Queremos calcular la funcin de distribucin acumulada (X<=9)
# P(Xk) = CDF de la distribucin geomtrica
p_acumulada = geom_defecto.cdf(k)
p_acumulada 
```

<mark>Escribe tu resumen y razonamiento de la solucin alcanzada en el apartado anterior aqu<mark>

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

```python
#3. Calcular las siguientes probabilidades $P(x4)$, $P(x=4)$, $P(x>4)$ y $P(x<4)$ e interpretar los resultados.

k = 4
p = 0.05
prob_le_4 = geom_defecto.cdf(k)
prob_eq_4 = geom_defecto.pmf(k)
prob_gt_4 = 1 - prob_le_4
prob_lt_4 = geom_defecto.cdf(k - 1)

prob_le_4, prob_eq_4, prob_gt_4, prob_lt_4
```

<mark>Escribe tu resumen y razonamiento de la solucin alcanzada en el apartado anterior aqu<mark>

```python
"""
4. Generar una muestra de tamao 1000 de una distribucin geomtrica siendo p=0.05 y 
luego calcular las frecuencias relativas pd.Series().count_values(normalize=True) y 
compararlas con las probabilidades tericas .pmf()
"""
import pandas as pd
from scipy.stats import geom

sample_size = 1000
p = 0.05

# Generar muestra
muestra = geom_defecto.rvs(size=sample_size)

# Calcular frecuencias relativas
freq_rel = pd.Series(muestra).value_counts(normalize=True).sort_index()

# Calcular probabilidades tericas
x = np.arange(1, freq_rel.index.max() + 1)
prob_teoricas = geom_defecto.pmf(x)

# Crear DataFrame para comparacin
df_comparacion = pd.DataFrame({
    'Frecuencia Relativa': freq_rel,
    'Probabilidad Terica': prob_teoricas[:len(freq_rel)]
})

print("Comparacin de las primeras 10 probabilidades:")
print(df_comparacion.head(10).round(4))

# Visualizacin
plt.figure(figsize=(10, 6))
plt.bar(freq_rel.index, freq_rel.values, alpha=0.5, label='Frecuencias Relativas', color='blue')
plt.plot(x, prob_teoricas, 'r-', label='Probabilidades Tericas')
plt.title('Comparacin entre Frecuencias Relativas y Probabilidades Tericas')
plt.xlabel('Nmero de pruebas hasta primer defecto')
plt.ylabel('Probabilidad')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Calcular algunas estadsticas descriptivas
print("\nEstadsticas descriptivas de la muestra:")
print(pd.Series(muestra).describe().round(2))
```

```python
#1. Qu distribucin de probabilidad sigue la variable aleatoria X? Proporciona el tipo de distribucin y sus parmetros.
1. La variable aleatoria X sigue una distribucin Binomial con parmetros:
* n = 100 (nmero de das/ensayos)
* p = 0.5 (probabilidad de xito/aumento)
* Justificacin:
    * Tenemos un nmero fijo de ensayos (100 das)
    * Cada ensayo es independiente
    * Solo hay dos resultados posibles cada da (sube o no sube)
    * La probabilidad de xito es constante (50%)
* Por tanto, X ~ B(100, 0.5)
```

```python
#2. Cul es la probabilidad de que el valor de las acciones aumente exactamente 50 veces en los 100 das? 
#Parmetrps
n = 100
p = 0.5

from scipy.stats import binom

binom_sube = binom(n=n, p=p)

prob_x_50 = binom_sube.pmf(50)
prob_x_50
```
