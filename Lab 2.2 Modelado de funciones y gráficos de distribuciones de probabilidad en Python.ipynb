### Ejercicio 1

Supongamos que gestionamos una pequeña empresa que vende un producto online. Basándonos en históricos, sabemos que, en promedio, el 20% de las personas que visitan nuestro sitio web realizan una compra. Ahora, en un día particular, tenemos 500 visitantes en el sitio. Asumir que la variable aleatoria $x$ (número de compras) sigue una distribución binomial, ya que tenemos un número fijo de ensayos (500 visitantes), cada ensayo es independiente, y cada ensayo resulta en éxito (compra) con una probabilidad del 20%.

1. ¿Qué distribución de probabilidad tiene $x$? Tenéis que justificar el tipo de distribución y sus parámetros.

2. ¿Cuál es la probabilidad de tener exactamente 100 compras en un día, dado que 500 personas visitan el sitio?

3. ¿Cuál es la probabilidad de tener más de 100 compras en un día, dado que 500 personas visitan el sitio?

4. ¿Cuál es la probabilidad de tener menos de 100 compras en un día, dado que 500 personas visitan el sitio?

5. Representar gráficamente la función de masa de la variable aleatoria $x$ e interpretar los resultados.

<div style="background-color:#F7DC6F"><b>Resolución Ejercicio 01</b></div>

<div style="background-color:#5DADE2"><b>Apartado 1</b></div>

```python
#1. ¿Qué distribución de probabilidad tiene $x$? Tenéis que justificar el tipo de distribución y sus parámetros.
La función a continuación es una función Binomial (X ~ B(500, 0.20))
Por los siguientes motivos:
* Tenemos un número fijo de ensayos (500)
* Cada ensayo es independiente
* Solo hay dos resultados posibles: compra o no compra
* La probabilidad de éxito es constante (20%)
```

<mark>Escribe tu resumen y razonamiento de la solución alcanzada en el apartado anterior aquí<mark>

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

```python
#2. ¿Cuál es la probabilidad de tener exactamente 100 compras en un día, dado que 500 personas visitan el sitio?
from scipy.stats import binom

#Parámetros
n = 500
p = 0.2

#Creamos la distribución binomial
bin_compras = binom(n=n, p=p)

#Queremos calcular la probabilidad de tener exactamente 100 compras en un día
prob_eq_100 = bin_compras.pmf(100)
print(f'La probabilidad de tener exactamente 100 compras en un día: {prob_eq_100}')
```

<mark>Escribe tu resumen y razonamiento de la solución alcanzada en el apartado anterior aquí<mark>

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

```python
#3. ¿Cuál es la probabilidad de tener más de 100 compras en un día, dado que 500 personas visitan el sitio?
from scipy.stats import binom

#Parámetros
n= 500
p = 0.2

#Creamos la distribución binomial
bin_compras = binom(n=n, p=p)

prob_h_100 = 1 - bin_compras.cdf(100)
print(f'La probabilidad de tener más de 100 compras en un día: {prob_h_100}')
```

<mark>Escribe tu resumen y razonamiento de la solución alcanzada en el apartado anterior aquí<mark>

<div style="background-color:#5DADE2"><b>Apartado 4</b></div>

```python
#4. ¿Cuál es la probabilidad de tener menos de 100 compras en un día, dado que 500 personas visitan el sitio?
from scipy.stats import binom

#Parámetros
n= 500
p= 0.2

#Creamos la distribución binmomial
bin_compras = binom(n=n, p=p)

prob_l_100 = bin_compras.cdf(99)
print(f'La probabilidad de tener menos de 100 compras en un día (es decir 99): {prob_l_100}')
```

<mark>Escribe tu resumen y razonamiento de la solución alcanzada en el apartado anterior aquí<mark>

<div style="background-color:#5DADE2"><b>Apartado 5</b></div>

```python
#Representar gráficamente la función de masa de la variable aleatoria e interpretar los resultados.
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
# Generar valores para la gráfica
x = bin_compras.rvs(size=n) # rango de valores para visualizar

#Funciónd e masa de probabilidad para todos los valores de x
pmf = bin_compras.pmf(x)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.bar(x, pmf, width=1, edgecolor="blue")

# Personalizar la gráfica
plt.title('Distribución Binomial B(500, 0.20)', fontsize=12)
plt.xlabel('Número de Compras (X)', fontsize=10)
plt.xlim(50, 150)
plt.ylabel('Probabilidad P(X = k)', fontsize=10)
plt.grid(True, alpha=0.3)
```

### Ejercicio 2

En una línea de producción de chips para ordenadores, se sabe que hay una probabilidad del 5% de que un chip sea defectuoso. Una máquina automática examina cada chip de manera secuencial, y la producción continúa hasta que se encuentra un chip defectuoso, momento en el cual la línea de producción se detiene para hacer ajustes y correcciones necesarias.

1. ¿Qué distribución de probabilidad consideramos en este caso? Tenéis que justificar el tipo de distribución y sus parámetros.

2. Calcular la probabilidad de que se necesiten probar $10$ chips para encontrar el primer chip defectuoso. 

3. Calcular las siguientes probabilidades $P(x≤4)$, $P(x=4)$, $P(x>4)$ y $P(x<4)$ e interpretar los resultados.
 
4. Generar una muestra de tamaño $1000$ de una distribución geométrica siendo $p=0.05$ y luego calcular las frecuencias relativas (`pd.Series().count_values(normalize=True)`) y compararlas con las probabilidades teóricas (`.pmf()`).

<div style="background-color:#F7DC6F"><b>Resolución Ejercicio 02</b></div>

<div style="background-color:#5DADE2"><b>Apartado 1</b></div>

```python
#1. ¿Qué distribución de probabilidad consideramos en este caso? Tenéis que justificar el tipo de distribución y sus parámetros.
La función a continuación es una función Geométrica con parámetro p= 0.05
Por los siguientes motivos:
* Contamos el número de ensayos hasta el primer éxito
* Cada ensayo es independiente
* La probabilidad de éxito es constante
* No hay límite en el número de ensayos
```

<mark>Escribe tu resumen y razonamiento de la solución alcanzada en el apartado anterior aquí<mark>

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

```python
#2. Calcular la probabilidad de que se necesiten probar 10 chips para encontrar el primer chip defectuoso. 
from scipy.stats import geom

#Parámetros
k = 9
p = 0.05

#Creamos la distribución geométrica
geom_defecto = geom(p=p)

#Queremos calcular la función de distribución acumulada (X<=9)
# P(X≤k) = CDF de la distribución geométrica
p_acumulada = geom_defecto.cdf(k)
p_acumulada 
```

<mark>Escribe tu resumen y razonamiento de la solución alcanzada en el apartado anterior aquí<mark>

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

```python
#3. Calcular las siguientes probabilidades $P(x≤4)$, $P(x=4)$, $P(x>4)$ y $P(x<4)$ e interpretar los resultados.

k = 4
p = 0.05
prob_le_4 = geom_defecto.cdf(k)
prob_eq_4 = geom_defecto.pmf(k)
prob_gt_4 = 1 - prob_le_4
prob_lt_4 = geom_defecto.cdf(k - 1)

prob_le_4, prob_eq_4, prob_gt_4, prob_lt_4
```

<mark>Escribe tu resumen y razonamiento de la solución alcanzada en el apartado anterior aquí<mark>

<div style="background-color:#5DADE2"><b>Apartado 4</b></div>

```python
"""
4. Generar una muestra de tamaño 1000 de una distribución geométrica siendo p=0.05 y 
luego calcular las frecuencias relativas pd.Series().count_values(normalize=True) y 
compararlas con las probabilidades teóricas .pmf()
"""
import pandas as pd
from scipy.stats import geom

sample_size = 1000
p = 0.05

# Generar muestra
muestra = geom_defecto.rvs(size=sample_size)

# Calcular frecuencias relativas
freq_rel = pd.Series(muestra).value_counts(normalize=True).sort_index()

# Calcular probabilidades teóricas
x = np.arange(1, freq_rel.index.max() + 1)
prob_teoricas = geom_defecto.pmf(x)

# Crear DataFrame para comparación
df_comparacion = pd.DataFrame({
    'Frecuencia Relativa': freq_rel,
    'Probabilidad Teórica': prob_teoricas[:len(freq_rel)]
})

print("Comparación de las primeras 10 probabilidades:")
print(df_comparacion.head(10).round(4))

# Visualización
plt.figure(figsize=(10, 6))
plt.bar(freq_rel.index, freq_rel.values, alpha=0.5, label='Frecuencias Relativas', color='blue')
plt.plot(x, prob_teoricas, 'r-', label='Probabilidades Teóricas')
plt.title('Comparación entre Frecuencias Relativas y Probabilidades Teóricas')
plt.xlabel('Número de pruebas hasta primer defecto')
plt.ylabel('Probabilidad')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Calcular algunas estadísticas descriptivas
print("\nEstadísticas descriptivas de la muestra:")
print(pd.Series(muestra).describe().round(2))
```

```python
"""
6. Generar una muestra de 1000 valores de una variable aleatoria que sigue una distribución de Poisson (poisson) 
con un parámetro de λ=3. Realizar un histograma de los valores generados y compara la distribución empírica con la teórica, 
justificando las similitudes o diferencias que observes.
"""
from scipy.stats import poisson

#Parámetro de la disctribución de Poisson
lam = 3

#Creamos la distribución de Poisson
poisson_aleat = poisson(mu=lam)

#Generamos la muestra
sample_size= 1000
poisson_sample = poisson_aleat.rvs(size=sample_size)

#Calcular las frecuencias teóricas
valores_x = np.arange(0, np.max(poisson_sample) + 1)
prob_teoricas = poisson.pmf(valores_x, mu=lam)

#Grafcar histograma y comparar
plt.figure(figsize=(10,6))
plt.hist(poisson_sample, bins=np.arange(0, np.max(poisson_sample) + 2) -0.5, density=True, alpha=0.6, label="Frecuencia Muestral")
plt.plot(valores_x, prob_teoricas, 'o-', color='red', label="Probabilidades Teóricas")
plt.title("Comparación entre Histograma Muestral y Distribución Teórica de Poisson")
plt.xlabel("Valores")
plt.ylabel("Frecuencia / Probabilidad")
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

```python
#1. ¿Qué distribución de probabilidad sigue la variable aleatoria X? Proporciona el tipo de distribución y sus parámetros.
1. La variable aleatoria X sigue una distribución Binomial con parámetros:
* n = 100 (número de días/ensayos)
* p = 0.5 (probabilidad de éxito/aumento)
* Justificación:
    * Tenemos un número fijo de ensayos (100 días)
    * Cada ensayo es independiente
    * Solo hay dos resultados posibles cada día (sube o no sube)
    * La probabilidad de éxito es constante (50%)
* Por tanto, X ~ B(100, 0.5)
```

```python
#2. ¿Cuál es la probabilidad de que el valor de las acciones aumente exactamente 50 veces en los 100 días? 
#Parámetrps
n = 100
p = 0.5

from scipy.stats import binom

binom_sube = binom(n=n, p=p)

prob_x_50 = binom_sube.pmf(50)
prob_x_50
```
