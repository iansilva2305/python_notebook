```python
"""
Definir una matriz x de ceros de 4 filas y 4 columnas. 
Tras ello sumar un 1 a cada nmero de la matriz y dividir el resultado entre 4. 
Tras definir la matriz x, calcular su transpuesta y obtener las dimensiones de la matriz traspuesta.
"""
import numpy as np

# 1. Definir matriz de ceros 4x4
x = np.zeros((4, 4))

# 2. Sumar 1 a cada elemento y dividir entre 4
x = (x + 1) / 4

# 3. Calcular la transpuesta
x_transpuesta = x.T

# 4. Obtener dimensiones de la transpuesta
dimensiones = x_transpuesta.shape

print("Matriz x:")
print(x)
print("\nMatriz transpuesta:")
print(x_transpuesta)
print("\nDimensiones de la matriz transpuesta:", dimensiones)
```

```python
"""
2. Calcular la suma y resta de cada fila de la matriz np.arange(1, 17).reshape(4, 4) usando la funcin np.apply_along_axis.
"""
# Importar las bibliotecas necesarias
import numpy as np

# Crear la matriz
matriz = np.arange(1, 17).reshape(4, 4)

# Definir las funciones para la suma y resta de cada fila
def suma_fila(fila):
    return np.sum(fila)

def resta_fila(fila):
    return np.subtract.reduce(fila)

# Aplicar las funciones a lo largo de cada fila usando np.apply_along_axis
suma_filas = np.apply_along_axis(suma_fila, axis=1, arr=matriz)
resta_filas = np.apply_along_axis(resta_fila, axis=1, arr=matriz)

# Imprimir los resultados
print("Matriz original:")
print(matriz)
print("\nSuma de cada fila:")
print(suma_filas)
print("\nResta de cada fila:")
print(resta_filas)
```

```python
"""
4. Con la matriz M, calcular su traspuesta, adjunta, el determinante y obtener la inversa de dicha matriz paso a paso. Explicarlo.
"""

# Definir la matriz como un DataFrame de pandas para mejor visualizacin
matriz = pd.DataFrame([[2, 1, 1, 0],
                       [4, 3, 3, 1],
                       [8, 7, 9, 5],
                       [6, 7, 9, 8]])

# 1. Calcular la traspuesta de la matriz
traspuesta = matriz.transpose()
print("Traspuesta de la matriz:")
print(traspuesta)

# 2. Calcular el determinante de la matriz
determinante = np.linalg.det(matriz.values)
print("\nDeterminante de la matriz:")
print(determinante)

# 3. Calcular la matriz adjunta
# La adjunta es la traspuesta de la matriz de cofactores
cofactores = np.zeros(matriz.shape)
for i in range(matriz.shape[0]):
    for j in range(matriz.shape[1]):
        # Crear la submatriz al eliminar la fila i y columna j
        submatriz = np.delete(np.delete(matriz.values, i, axis=0), j, axis=1)
        # Calcular el cofactor
        cofactores[i, j] = (-1) ** (i + j) * np.linalg.det(submatriz)
adjunta = pd.DataFrame(cofactores).transpose()
print("\nMatriz adjunta:")
print(adjunta)

# 4. Calcular la inversa de la matriz
if determinante != 0:
    inversa = adjunta / determinante
    inversa = pd.DataFrame(inversa)
    print("\nInversa de la matriz:")
    print(inversa)
else:
    print("\nLa matriz no tiene inversa (determinante = 0).")
```
